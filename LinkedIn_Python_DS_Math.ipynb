## Nonparametric Correlation Analysis

```python
import scipy
from scipy.stats import spearmanr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 14,7
```

```python
# Using mtcars data set
cars.head()
```

```python
sb.pairplot(cars)
```

```python
x = cars[['cyl','vs','am','gear']]
sb.pairplot(x)
```

```python
# Isolating variables
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coefficient, p_value = spearmanr(cyl,vs)

print('spearman Rank Correlation Coeffecient %0.3f' %(spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl,am)

print('spearman Rank Correlation Coeffecient %0.3f' %(spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl,gear)

print('spearman Rank Correlation Coeffecient %0.3f' %(spearmanr_coefficient))
```

### Chi-square test for indepenedence

```python
table = pd.crosstab(cyl,am)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl,vs)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```

```python
#Test question
pd.crosstab(vs, cyl)
```

```python
table = pd.crosstab(cyl,gear)

from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```
